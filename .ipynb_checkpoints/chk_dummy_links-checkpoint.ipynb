```python
import pandas as pd
import geopandas as gpd
import numpy as np
import os, sys 
```

```python
hwy_network = gpd.read_file(r'M:\Application\Model One\Networks\TM1_2015_Base_Network\shapefile\freeflow_links.shp')
```

```python
print(hwy_network.shape[0])
print(list(hwy_network))
display(hwy_network.head())
```

```python
links_df = hwy_network.copy()
```

## calculations from Wrangler/TransitNetwork.py to find HOV lanes' corresponding GP lanes

```python
# filter out HOV and express lane links
hov_links_df        = links_df.loc[ (links_df.USE == 2)|(links_df.USE==3) ]
el_links_df         = links_df.loc[ links_df.TOLLCLASS >= 11 ]
notruck_links_df    = links_df.loc[ (links_df.TOLLCLASS==0) & (links_df.USE==4)]        
gp_links_df         = links_df.loc[ (links_df.USE==1)&((links_df.FT<=3)|(links_df.FT==5)|(links_df.FT==7)|(links_df.FT==8)|(links_df.FT==10))]
gp_notruck_links_df = pd.concat([gp_links_df, notruck_links_df])
dummy_links_df      = links_df.loc[ links_df.FT==6 ]

print("Found {} hov links, {} express lane links and {} general purpose links".format(
                     len(hov_links_df), len(el_links_df), len(gp_notruck_links_df)))
```

```python
# dummy B -> hov A, a_GP1 will be the first point of dummy access link
hov_group1_df  = pd.merge(left=hov_links_df, right=dummy_links_df[["A","B"]],
                              how="inner", left_on=["A"], right_on=["B"], suffixes=["","_GP1"]).drop(columns="B_GP1")
display(hov_group1_df)

# hov B -> dummy A, b_GP2 will be the second point of dummy egress link
hov_group1_df  = pd.merge(left=hov_group1_df, right=dummy_links_df[["A","B"]],
                              how="inner", left_on=["B"], right_on=["A"], suffixes=["","_GP2"]).drop(columns="A_GP2")
display(hov_group1_df)

# merge to the full GP links for complete info
hov_group1_df  = pd.merge(left=hov_group1_df, right=gp_notruck_links_df, how="inner",
                              left_on=["A_GP1", "B_GP2"], right_on=["A","B"], suffixes=["","_GP_through_dummy"]).drop(columns=["A_GP1","B_GP2"])
display(hov_group1_df)
WranglerLogger.debug("Found general purpose links for {} out of {} hov links: \n{}".format(
                     len(hov_group1_df), len(hov_links_df), hov_group1_df.head()))
```

```python
# Note which links don't have GP equivalents
hov_unmatched_df = pd.merge(left=hov_links_df, right=hov_group1_df[["A","B","A_GP_through_dummy","B_GP_through_dummy"]], how="left")
hov_unmatched_df = hov_unmatched_df.loc[hov_unmatched_df.A_GP_through_dummy.isnull()]
display(hov_unmatched_df)
```

## second round of matching - identify trangle situations

```python
# find when an HOV lane shares A node with a GP lane and form an trangle

hov_unmatched_df = hov_unmatched_df.merge(
    gp_notruck_links_df[['A', 'B']].rename(columns={
        'A': 'A_share_with_GP',
        'B': 'B_GP_trangle'}),
    left_on='A',
    right_on='A_share_with_GP',
    how='left')
```

```python
# find when an HOV lane shares B node with a GP lane and form an trangle

hov_unmatched_df = hov_unmatched_df.merge(
    gp_notruck_links_df[['A', 'B']].rename(columns={
        'A': 'A_GP_trangle',
        'B': 'B_share_with_GP'}),
    left_on='B',
    right_on='B_share_with_GP',
    how='left')
```
